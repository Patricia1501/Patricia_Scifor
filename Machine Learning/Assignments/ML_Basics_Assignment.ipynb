### <b>How machine learning algorithms work?</b><br>
Machine Learning works in the following manner.<br>

A machine learning algorithm works by learning patterns and relationships from data to make predictions or decisions without being explicitly programmed for each task. Heres a simplified overview of how a typical machine learning algorithm works:<br>

<b>1. Data Collection:</b><br>
First, relevant data is collected or curated. This data could include examples, features, or attributes that are important for the task at hand, such as images, text, numerical data, etc.<br>

<b>2. Data Preprocessing:</b><br>
Before feeding the data into the algorithm, it often needs to be preprocessed. This step may involve cleaning the data (handling missing values, outliers), transforming the data (normalization, scaling), and splitting it into training and test sets.<br>

<b>3. Choosing a Model:</b><br>
Depending on the task (e.g., classification, regression, clustering), a suitable machine learning model is chosen. Examples include decision trees, neural networks, support vector machines, and more advanced models like deep learning architectures.<br>

<b>4. Training the Model:<br></b>
The selected model is trained using the training data. During training, the algorithm learns patterns and relationships in the data. This involves adjusting model parameters iteratively to minimize the difference between predicted outputs and actual outputs (labels or targets) in the training data.<br>

<b>5. Evaluating the Model:</b><br>
Once trained, the model is evaluated using the test data to assess its performance. Metrics such as accuracy, precision, recall, or mean squared error are used to evaluate how well the model generalizes to new, unseen data.<br>

<b>6. Fine-tuning:</b><br>
Models may be fine-tuned by adjusting hyperparameters (parameters that are not directly learned during training, like learning rate or number of hidden layers in a neural network) to improve performance.<br>

<b>7. Prediction or Inference:</b><br>
Finally, the trained model is used to make predictions or decisions on new data. This process involves applying the learned patterns to new inputs to generate outputs, such as class labels in classification tasks or numerical values in regression tasks.<br>

### <b>Machine Learning lifecycle:</b><br>

The lifecycle of a machine learning project involves a series of steps that include:

<b>1. Study the Problems:</b><br>
The first step is to study the problem. This step involves understanding the business problem and defining the objectives of the model.

<b>2. Data Collection:</b><br>
When the problem is well-defined, we can collect the relevant data required for the model. The data could come from various sources such as databases, APIs, or web scraping.

<b>3. Data Preparation:</b><br>
When our problem-related data is collected. then it is a good idea to check the data properly and make it in the desired format so that it can be used by the model to find the hidden patterns. This can be done in the following steps:<br>
i)Data cleaning<br>
ii)Data Transformation<br>
iii)Explanatory Data Analysis and Feature Engineering<br>
iv)Split the dataset for training and testing.<br>

<b>4. Model Selection:</b><br>
The next step is to select the appropriate machine learning algorithm that is suitable for our problem. This step requires knowledge of the strengths and weaknesses of different algorithms. Sometimes we use multiple models and compare their results and select the best model as per our requirements.<br>

<b>5. Model building and Training:</b><br>
After selecting the algorithm, we have to build the model.
In the case of traditional machine learning building mode is easy it is just a few hyperparameter tunings.<br>
In the case of deep learning, we have to define layer-wise architecture along with input and output size, number of nodes in each layer, loss function, gradient descent optimizer, etc.
After that model is trained using the preprocessed dataset.<br>

<b>6. Model Evaluation:</b><br>
Once the model is trained, it can be evaluated on the test dataset to determine its accuracy and performance using different techniques. like classification report, F1 score, precision, recall, ROC Curve, Mean Square error, absolute error, etc.<br>

<b>7. Model Tuning:</b><br>
Based on the evaluation results, the model may need to be tuned or optimized to improve its performance. This involves tweaking the hyperparameters of the model.<br>

<b>8. Deployment:</b><br>
Once the model is trained and tuned, it can be deployed in a production environment to make predictions on new data. This step requires integrating the model into an existing software system or creating a new system for the model.<br>

<b>9. Monitoring and Maintenance:</b><br>
Finally, it is essential to monitor the models performance in the production environment and perform maintenance tasks as required. This involves monitoring for data drift, retraining the model as needed, and updating the model as new data becomes available.

### <b><u>1. Supervised Machine Learning:</b><br></u>
<b>Supervised learning</b> is a type of machine learning in which the algorithm is trained on the labeled dataset. It learns to map input features to targets based on labeled training data. In supervised learning, the algorithm is provided with input features and corresponding output labels, and it learns to generalize from this data to make predictions on new, unseen data.

<b>Example for Supervised Learning:</b><br>
a) We get a lot of emails every day. Some are important, and some are just spam. We want our email service to automatically move spam emails to the spam folder and keep important ones in our inbox.

b) Imagine you have a friend who is trying to sell their house, and they want to know how much it might sell for. You decide to use supervised learning to help predict the price.

There are two main types of supervised learning:

<b><u>Regression:</b></u><br> <b>Regression</b> is a type of supervised learning where the algorithm learns to predict continuous values based on input features. The output labels in regression are continuous values, such as stock prices, and housing prices.

The different regression algorithms in machine learning are: Linear Regression, Polynomial Regression, Ridge Regression, Decision Tree Regression, Random Forest Regression, Support Vector Regression, etc.<br>

<b><u>Classification:</b></u><br> <b>Classification</b> is a type of supervised learning where the algorithm learns to assign input data to a specific category or class based on input features. The output labels in classification are discrete values. Classification algorithms can be binary, where the output is one of two possible classes, or multiclass, where the output can be one of several classes.

The different Classification algorithms in machine learning are: Logistic Regression, Naive Bayes, Decision Tree, Support Vector Machine (SVM), K-Nearest Neighbors (KNN), etc

### <b><u>2. Unsupervised Machine Learning:</b></u><br>
<b>Unsupervised learning</b> is a type of machine learning where the algorithm learns to recognize patterns in data without being explicitly trained using labeled examples. The goal of unsupervised learning is to discover the underlying structure or distribution in the data.

<b>Example for Unsupervised Learning:</b><br>
a) Imagine you run an online store and you want to understand your customers better to target them with personalized marketing. However, you don't have any predefined labels like "high spender" or "bargain hunter" for each customer. You just have raw data about their behavior.<br>

b) You use a movie streaming service like Netflix. The service wants to recommend movies to you that you are likely to enjoy, but they dont have labels like liked or disliked for every user and every movie.<br>


There are two main types of unsupervised learning:

<b>Clustering:</b><br> <b>Clustering</b> algorithms group similar data points together based on their characteristics. The goal is to identify groups, or clusters, of data points that are similar to each other, while being distinct from other groups. Some popular clustering algorithms include K-means, Hierarchical clustering, and DBSCAN.<br>

<b>Dimensionality reduction:</b><br> <b>Dimensionality reduction</b> algorithms reduce the number of input variables in a dataset while preserving as much of the original information as possible. This is useful for reducing the complexity of a dataset and making it easier to visualize and analyze. Some popular dimensionality reduction algorithms include Principal Component Analysis (PCA), t-SNE, and Autoencoders.

### <b><u>3. Reinforcement Machine Learning:</b></u><br>
<b>Reinforcement learning</b> is a type of machine learning where an agent learns to interact with an environment by performing actions and receiving rewards or penalties based on its actions. The goal of reinforcement learning is to learn a policy, which is a mapping from states to actions, that maximizes the expected cumulative reward over time.

<b>Example for Reinforcement Learning:</b><br>
a) Imagine you're developing a self-driving car. The car needs to learn how to drive safely and efficiently on its own, making decisions like when to stop, turn, or change lanes.<br>

b) You want to train your dog to fetch a ball. The dog needs to learn that fetching the ball and bringing it back to you is a good thing.<br>

There are two main types of reinforcement learning:

<b>Model-based reinforcement learning:</b><br> In model-based reinforcement learning, the agent learns a model of the environment, including the transition probabilities between states and the rewards associated with each state-action pair. The agent then uses this model to plan its actions in order to maximize its expected reward. Some popular model-based reinforcement learning algorithms include Value Iteration and Policy Iteration.<br>

<b>Model-free reinforcement learning:</b> In model-free reinforcement learning, the agent learns a policy directly from experience without explicitly building a model of the environment. The agent interacts with the environment and updates its policy based on the rewards it receives. Some popular model-free reinforcement learning algorithms include Q-Learning, SARSA, and Deep Reinforcement Learning.

## <b><i>CLASSIFICATION v/s REGRESSION v/s CLUSTERING</i></b>:

Machine Learning (ML) has revolutionized the way we analyze and interpret data. Among its many applications, classification, regression, and clustering are fundamental techniques that allow us to uncover patterns and make predictions based on data characteristics.


maxresdefault.jpg


#### <b><u>Classification: Unraveling Unknown Data Points</u></b><br>
Classification is a vital aspect of ML that involves assigning unknown data points to predefined classes. When we have a binary classification problem with two classes (0 and 1), the task is to determine which class the data point belongs to. On the other hand, multi-class classification deals with scenarios where the data point can be assigned to one of several classes (1, 2, 3, 4, 5, 6, 7, 8, 9).

#### <b><u>Regression: Predicting Real Values</b></u><br>
In regression, we aim to predict real values based on the features present in the training data. Unlike classification, where we assign classes, regression focuses on estimating continuous and numerical values. The result of regression is a value on the real number line that represents the predicted outcome based on the given features.

#### <b><u>Clustering: Unveiling Data Similarities</b></u><br>
Unlike classification and regression, clustering does not involve assigning predefined labels or predicting values. Instead, clustering groups data points based on the similarity of their features. The goal is to identify inherent patterns and relationships within the data. Clustering is particularly useful when we have unlabeled data and want to discover hidden structures.

#### <b>Limitations of Machine Learning:-</b><br>
a) The primary challenge of machine learning is the lack of data or the diversity in the dataset.<br>
b) A machine cannot learn if there is no data available. Besides, a dataset with a lack of diversity gives the machine a hard time. <br>
c) A machine needs to have heterogeneity to learn meaningful insight. <br>
d) It is rare that an algorithm can extract information when there are no or few variations.<br>
e) It is recommended to have at least 20 observations per group to help the machine learn. This constraint leads to poor evaluation and prediction.<br>
