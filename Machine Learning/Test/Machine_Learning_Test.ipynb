```python
# Checking for missing values
df.isnull().sum()
```

```python
# Bar Chart:
avg_scores = df.groupby('test preparation course')[['math score']].mean() # Changed tuple to list
avg_scores.plot(kind='bar', figsize=(10, 6))
plt.title('Average Scores in maths')
plt.xlabel('Test Preparation Course')
plt.ylabel('Average Scores')
plt.legend(['Math Score'])
plt.show()
```

```python
# Pie Chart:
gender_counts = df['gender'].value_counts()
gender_counts.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightcoral'], figsize=(8, 8))
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
# Histogram:
df['reading score'].plot(kind='hist', bins=20, color='skyblue', edgecolor='black', figsize=(10, 6))
plt.title('Distribution of Reading Scores')
plt.xlabel('Reading Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Line Chart:
avg_scores_by_education = df.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean() # Changed tuple to list
avg_scores_by_education.plot(kind='line', marker='o', figsize=(10, 6))
plt.title('Average Scores by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Average Scores')
plt.legend(['Math Score', 'Reading Score', 'Writing Score'])
plt.show()
```
